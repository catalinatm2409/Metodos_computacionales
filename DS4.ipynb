```python
url = "/content/drive/MyDrive/Colab Notebooks/Datasets/diabetes_binary_5050split_health_indicators_BRFSS2015.csv"

import pandas as pd

df_diabetes = pd.read_csv(url)

df_diabetes.head()
```

```python
df_diabetes_feat = df_diabetes.drop(['Diabetes_binary'], axis=1)
df_diabetes_class = df_diabetes['Diabetes_binary']

df_diabetes_feat.shape, df_diabetes_class.shape
```

```python
from sklearn.model_selection import train_test_split

X_train0, X_test0, y_train0, y_test0 = train_test_split(df_diabetes_feat, df_diabetes_class, random_state = 0)
```

```python
X_train0.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train0, y_train0)

print("Score de X_train: {:.2f}".format(knn.score(X_train0, y_train0)))
print("Score de X_test: {:.2f}".format(knn.score(X_test0, y_test0)))
```

```python
from sklearn.metrics import confusion_matrix

pred_knn = knn.predict(X_test0)

confusion0 = confusion_matrix(y_test0, pred_knn)

print("Matriz de confusiÃ³n\n{}".format(confusion0))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Accuracy: {:.2f}'.format(accuracy_score(y_test0, pred_knn)))
print('Precision: {:.2f}'.format(precision_score(y_test0, pred_knn)))
print('Recall: {:.2f}'.format(recall_score(y_test0, pred_knn)))
print('F1: {:.2f}'.format(f1_score(y_test0, pred_knn)))
```

```python
for n in range(1,5):
  knn = KNeighborsClassifier(n_neighbors=n)
  knn.fit(X_train0, y_train0)
  print("Score de X_train: {:.2f}".format(knn.score(X_train0, y_train0)))
  print("Score de X_test: {:.2f}".format(knn.score(X_test0, y_test0)))
```

```python
knn3 = KNeighborsClassifier(n_neighbors=3)

knn3.fit(X_train0, y_train0)

pred_knn3 = knn3.predict(X_test0)

confusion3 = confusion_matrix(y_test0, pred_knn3)

print("Matriz de confusiÃ³n\n{}".format(confusion3))
```

```python
print('Accuracy: {:.2f}'.format(accuracy_score(y_test0, pred_knn3)))
print('Precision: {:.2f}'.format(precision_score(y_test0, pred_knn3)))
print('Recall: {:.2f}'.format(recall_score(y_test0, pred_knn3)))
print('F1: {:.2f}'.format(f1_score(y_test0, pred_knn3)))
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree

Decision_tree = DecisionTreeClassifier().fit(X_train0, y_train0)

Decision_tree.score(X_train0, y_train0), Decision_tree.score(X_test0, y_test0)
```

```python
pred_tree = Decision_tree.predict(X_test0)

confusion_t = confusion_matrix(y_test0, pred_tree)

print("Matriz de confusiÃ³n\n{}".format(confusion_t))
```

```python
listtr1 = []
listte1 = []

for n in range(2,11):
  Decision_tree_n = DecisionTreeClassifier(max_depth=n).fit(X_train0, y_train0)
  listtr1.append(Decision_tree_n.score(X_train0, y_train0))
  listte1.append(Decision_tree_n.score(X_test0, y_test0))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(range(2,11), listtr1, color = 'blue')
plt.scatter(range(2,11), listte1, color = 'green')
```

```python
pred_tree6 = Decision_tree6.predict(X_test0)

confusion_t6 = confusion_matrix(y_test0, pred_tree6)

print("Matriz de confusiÃ³n\n{}".format(confusion_t6))
```

```python
from sklearn.naive_bayes import GaussianNB

diab_bayes = GaussianNB().fit(X_train0, y_train0)

diab_bayes.score(X_train0, y_train0), diab_bayes.score(X_test0, y_test0)
```

```python
pred_nb = diab_bayes.predict(X_test0)

confusion_nb = confusion_matrix(y_test0, pred_nb)

print("Matriz de confusiÃ³n\n{}".format(confusion_nb))
```
